
***2.  Import Factorisation class from the package***



```python
from  factodiagrams.preprocess.Factorisation import *
from IPython.display import display,Image
```

**3.  Draw diagram of number function**

```python
int_number=70
Factorization().draw_factor(int_number,"prime_factors",True,True)
Image(str(int_number)+"'s_Diagram.png", width=400)
```

***4. Draw The poster***

```python
int_number=70
Factorization().draw_factor_poster(list(range(1,70)),"prime_factors",5)
Image("poster.png",width=700)
```
